```python
import math

def calculate_cyclone_efficiency(particle_density, particle_diameter, inlet_velocity, barrel_diameter, num_turns):
    # Constants
    gravitational_acceleration = 9.81  # m/s^2
    air_density = 1.22  # kg/m^3 (approximate value)
    dynamic_viscosity = 1.8e-6  # Pa*s (approximate value)
    pi = 3.14

    # Calculate cut diameter
    cut_diameter = math.sqrt((9 * dynamic_viscosity * barrel_diameter) / (
            2 * pi * num_turns * inlet_velocity * (particle_density - air_density)))

    # Calculate overall removal efficiency
    overall_efficiency = 1-((cut_diameter/particle_diameter)**2)

    return cut_diameter, overall_efficiency

# Example usage:
particle_density = float(input("Enter particle density (in kg/m^3): "))
particle_diameter = float(input("Enter average particle diameter (in meters): "))
inlet_velocity = float(input("Enter inlet velocity (in m/s): "))
barrel_diameter = float(input("Enter barrel diameter (in meters): "))
num_turns = float(input("Enter number of turns: "))

cut_diameter, overall_efficiency = calculate_cyclone_efficiency(particle_density, particle_diameter, inlet_velocity, barrel_diameter, num_turns)

print("Cut Diameter:", cut_diameter)
print("Overall Efficiency:", overall_efficiency)
```
